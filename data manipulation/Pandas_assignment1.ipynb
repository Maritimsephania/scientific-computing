```python
# Aggregate average scores by borough
avg_scores = data.groupby('borough')[['average_math', 'average_reading', 'average_writing']].mean()
avg_scores
```

```python
# Plotting
avg_scores.plot(kind='bar', figsize=(10, 6), width=0.8)

# Customize the chart
plt.title('Average Scores by Borough', fontsize=14)
plt.xlabel('borough', fontsize=12)
plt.ylabel('Average Score', fontsize=12)
plt.legend(title='Score Type')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
average_tested = data.groupby('borough')[['percent_tested']].mean()
average_tested
```

```python
# Plotting
average_tested.plot(kind='bar', figsize=(10, 6), width=0.8)

# Customize the chart
plt.title('Average Scores by Borough', fontsize=14)
plt.xlabel('borough', fontsize=12)
plt.ylabel('average_tested', fontsize=12)
plt.legend(title='Score Type')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
data.plot(kind='scatter', x='average_writing', y='average_math', figsize=(10, 6))
plt.title('Average Math Scores vs. average_writing Scores', fontsize=14)
plt.xlabel('average_writing', fontsize=12)
plt.ylabel('average_math', fontsize=12)
plt.grid(True)
plt.show()
```

```python
data[['average_writing', 'average_reading']].plot(kind='hist', bins=10, alpha=0.6, figsize=(10, 6), legend=True)
plt.title('Distribution of Average Writing and Reading Scores')
plt.xlabel('Score')
plt.ylabel('Number of Schools')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#compute the correlation matrix
corr = data[['average_math', 'average_reading', 'average_writing']].corr()  
corr
```

```python
#plot the correlation matrix
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
